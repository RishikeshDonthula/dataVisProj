```python
df15sp['W'] = df15sp.apply(lambda x : re.match('^[\d]*', x['W–L'])[0], axis = 1)
df15sp['L'] = df15sp.apply(lambda x : re.search('(?<=–).*', x['W–L'])[0], axis = 1)
df15sp['W/L'] = df15sp.apply(lambda x: int(x['W'])/int(x['L']), axis=1)
```

```python
stats = pd.merge(df15sp[['Team', 'W/L']], df15su[['Team', 'W/L']], how = 'outer', on = "Team")
stats = stats.rename({'W/L_x':'2015 spring',
              'W/L_y':'2015 summer'}, axis = 1)
stats = pd.merge(stats, df16sp[['Team', 'W/L']], how = 'outer', on = "Team")
stats = stats.rename({'W/L': '2016 spring'}, axis = 1)

stats = pd.merge(stats, df16su[['Team', 'W/L']], how = 'outer', on = "Team")
stats = stats.rename({'W/L': '2016 summer'}, axis = 1)


stats = pd.merge(stats, df17su[['Team', 'W/L']], how = 'outer', on = "Team")
stats = stats.rename({'W/L': '2017 summer'}, axis = 1)

stats = pd.merge(stats, df18sp[['Team', 'W/L']], how = 'outer', on = "Team")
stats = stats.rename({'W/L': '2018 spring'}, axis = 1)

stats = pd.merge(stats, df18su[['Team', 'W/L']], how = 'outer', on = "Team")
stats = stats.rename({'W/L': '2018 summer'}, axis = 1)

stats = pd.merge(stats, df19sp[['Team', 'W/L']], how = 'outer', on = "Team")
stats = stats.rename({'W/L': '2019 spring'}, axis = 1)

stats = pd.merge(stats, df19su[['Team', 'W/L']], how = 'outer', on = "Team")
stats = stats.rename({'W/L': '2019 summer'}, axis = 1)

stats = pd.merge(stats, df20sp[['Team', 'W/L']], how = 'outer', on = "Team")
stats = stats.rename({'W/L': '2020 spring'}, axis = 1)

stats = pd.merge(stats, df20su[['Team', 'W/L']], how = 'outer', on = "Team")
stats = stats.rename({'W/L': '2020 summer'}, axis = 1)





stats = stats.set_index('Team')
```

```python
display(stats)
```

```python
stats = stats.loc[['Team SoloMid', 'Cloud9', 'Counter Logic Gaming', 'Team Liquid']]
```

```python
stats = stats.T
stats.plot(figsize = (15, 10), title = "Winrates of 4 Major Teams in the League Championship Series 2015-2020", \
           xlabel = 'Seasons (excluding 2017 spring)', ylabel = 'Winrate (W/L)', grid = True)
plt.xticks([i for i in range(len(stats.index.tolist()))], stats.index.tolist())

```
